```python
# Sales Data Exploratory Data Analysis

This notebook analyzes retail sales data to uncover trends and business insights.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
import pandas as pd

df = pd.read_csv('superstore.csv (1).csv')
df.head()
```

```python
# Shape of dataset
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

monthly_sales.plot(figsize=(12,5))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum)
plt.title('Total Profit by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Clean column names (remove extra spaces)
df.columns = df.columns.str.strip()

df.columns
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum)
plt.title('Total Profit by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
```

```python
category_sales = (
    df.groupby('Category', as_index=False)['Sales']
    .sum()
)

plt.figure(figsize=(8,5))
sns.barplot(data=category_sales, x='Category', y='Sales')
plt.title('Total Sales by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
region_sales = (
    df.groupby('Region', as_index=False)['Sales']
    .sum()
)

plt.figure(figsize=(8,5))
sns.barplot(data=region_sales, x='Region', y='Sales')
plt.title('Total Sales by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales = (
    df.groupby(df['Order Date'].dt.to_period('M'))['Sales']
    .sum()
)

monthly_sales.plot(figsize=(12,5))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

Final Insights

Sales demonstrate a clear seasonal pattern, with peaks during certain months each year.

Overall sales show a positive upward trend over time, indicating business growth.

Time-based analysis helps identify high-performing periods for strategic planning.
